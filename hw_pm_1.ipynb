#  Как бы Вы организовали аналитику по данному продукту?
Исходя из выручки:
4 потока в год, по ~20 человек, каждый по 170000 заплатил. Выручка =4*20*170.000=13,6 млн руб.
Рентабельность ~30% => Прибыль 4 млн. (Доходы - Расходы с учетом налогов)
Если мы выделим Senior Data science на проект, его средняя зарплата = 250 тр/мес(net) 
или 250*1.5=375 тр./мес. (с учетом всех налогов)*12 мес. =4,5 млн. руб./год 
Вывод: В текущий момент проект не имеет потенциала, чтобы нанять на полный день Senior Data scientist  

Значит на программу должны использоваться текущие преподаватели или аналитики, которые работают на остальных программах курса

1. Скорость роста продаж. На стадии роста компании (по И.Адизесу), ключевым моментом является увеличение объемов продаж. При этом быстро отрабатываются бизнес-процессы и снижаются удельные затраты.
2. NPS -> В долгосрочной перспективе DS сообщество активно общается между собой, если показатель будет низким, даже ценой не купить будет студентов.
